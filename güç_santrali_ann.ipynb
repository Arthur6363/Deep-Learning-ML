```python
import numpy as np
import pandas as pd
import tensorflow as tf
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
df = pd.read_csv("folds.csv")
```

```python
df.head()
```

```python
x = df.drop(["PE"], axis=1)
y= df["PE"]
```

```python

x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=63, test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_trainc = sc.fit_transform(x_train)
x_testc = sc.transform(x_test)
```

```python
ann = Sequential()
```

```python
ann.add(Dense(units=6, activation="relu", input_dim=4))
```

```python
ann.add(Dense(units=6, activation="relu"))
```

```python
ann.add(Dense(units=1))
```

```python
ann.compile(optimizer = "adam", loss="mean_squared_error",metrics=["accuracy"])
```

```python
ann.fit(x_trainc,y_train,epochs=100)
```

```python
ypred = ann.predict(x_testc)
```
